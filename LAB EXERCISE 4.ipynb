# Information Gain Coding Part


```python
from math import log2
```

```python
# defining the entropy func for the split in the dataset taking care of no impurity part
def entropy(class0, class1):
    if class0 and class1 !=0:
    
        return -(class0 * log2(class0) + class1 * log2(class1))
    else:
        return 0
```

# Taking inputs for different class values of child and parent
# Doing this for outlook instance

```python
# split of the main dataset
class0 = 9/14
class1 = 5/14

# calculate entropy before the change
s_entropy = entropy(class0, class1)
print('Dataset Entropy: %.3f' % s_entropy)
```

```python
# split 1 (split via value1)
s1_class0 = 4/4
s1_class1 = 0/4

# calculate the entropy of the first group
s1_entropy = entropy(s1_class0, s1_class1)
print('Group1 Entropy: %.3f' % s1_entropy)


# split 2  (split via value2)
s2_class0 = 2/5
s2_class1 = 3/5

# calculate the entropy of the second group
s2_entropy = entropy(s2_class0, s2_class1)
print('Group2 Entropy: %.3f' % s2_entropy)

# split 3  (split via value3)
s3_class0 = 3/5
s3_class1 = 2/5

# calculate the entropy of the second group
s3_entropy = entropy(s3_class0, s3_class1)
print('Group2 Entropy: %.3f' % s3_entropy)
```

# Calculate the information gain

```python
gain = s_entropy - (4/10* s1_entropy + 5/14 * s2_entropy + 5/14 * s3_entropy)
print('Information Gain: %.3f ' % gain)
```
